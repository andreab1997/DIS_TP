```python
import Integration as Int
import Initialize as Ini
import lhapdf
import numpy as np
import matplotlib.pyplot as plt
import MassiveCoeffFunc as mcf
import TildeCoeffFunc as tcf
import scipy.integrate as integrate
from mpl_toolkits import mplot3d
%matplotlib widget
```

```python
Int.Initialize_all()
```

```python
eps = 0.5
mine = lhapdf.mkPDF("MyPDF_mub=mb_nlo")
thre = [mine.quarkMass(5)*ratio for ratio in [0.5,1.,2.0]]
Qlogmin = np.log10(1.)
Qlogmax = np.log10(150.)
Qlog = np.linspace(Qlogmin,Qlogmax,200)
Qcommon = pow(10,Qlog)
Qsing = [np.linspace(thr-eps,thr+eps,5) for thr in thre]
Q = np.sort(np.concatenate((Qcommon,Qsing[0],Qsing[1],Qsing[2])))
```

```python
res = [mcf.Cg_2_m_reg(0.1,q,[mine.quarkMass(5)]) for q in Q]
```

```python
Q_arr = np.linspace(2,75,130)
Z_arr = np.geomspace(0.8,1,130)
Q,Z = np.meshgrid(Q_arr, Z_arr)
ax = plt.figure().add_subplot(projection='3d')
C = np.array([[tcf.Cg_2_til_reg(z,q,[mine.quarkMass(5)]) for q in Q_arr] for z in Z_arr])
D = np.array([[tcf.Cg_2_til_reg(z,q,[mine.quarkMass(5)], grids=False)[0] for q in Q_arr] for z in Z_arr])
ax.plot_surface(Q, Z, C, cmap='autumn', cstride=2, rstride=2)
ax.plot_surface(Q, Z, D, cmap='inferno', cstride=2, rstride=2)
plt.show()

```

```python
def fakecg2(q, thre):
    if q < thre:
        return 0
    return -1e1*(np.log(q)-1)**2
```

```python
ax1 = plt.figure().add_subplot()
resfake = [fakecg2(q,3.2) for q in Q]
ax1.plot(Q,res,Q,resfake)
ax1.set_xscale("log")
plt.show()
```

```python
pdf = lhapdf.mkPDF("MyPDF_4F_nnlo")
```

```python
ax2 = plt.figure().add_subplot()
result = [integrate.quad(lambda z: mcf.Cg_2_m_reg(z,q,[mine.quarkMass(5)])*pdf.xfxQ2(21,0.1*(1.0/z),q**2),0.1,1.0)[0] for q in Q]
ax2.plot(Q,result)
ax2.set_xscale("log")
plt.show()
```
